```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8)

dataframe= pd.read_csv(r'C:\Users\17742\Desktop\Data Projects\best cities for a workation.csv')
```

```python
dataframe.head()
```

```python
##Hypothesis: Price of a 1 bedroom apartment per month would have a postive correlation with price of a cup of coffee.
## More expensive rent usually means more expensive costs of goods and services 

plt.scatter(x= dataframe['Accommodation: Average price of 1 bedroom apartment per month'], y= dataframe['Caffeine: Average price of buying a coffee'])

sns.regplot(x='Accommodation: Average price of 1 bedroom apartment per month', y='Caffeine: Average price of buying a coffee', data= dataframe,line_kws={"color":"black"})
```

```python
##Hypothesis: Price of Beer and Price of Coffee have a postive correlation. Because both have a similar buying demographic.

plt.scatter(x= dataframe['After-work drinks: Average price for 2 beers in a bar'], y= dataframe['Caffeine: Average price of buying a coffee'], c="brown")

plt.title("Cost of 2 Beers vs. Coffee")

sns.regplot(x='After-work drinks: Average price for 2 beers in a bar', y='Caffeine: Average price of buying a coffee', data= dataframe,line_kws={"color":"green"})
```

```python
##Hypothesis: I thought the more annual sunshine a place had the more instagram pictures would have been taken. I assumed
## people prefer warmer areas, but with a log correlation it seems people hashtag all around the world with low correlation

plt.scatter(x=dataframe["Climate: Average number of sunshine hours"], y = dataframe["Instagramability: Number of photos with #"], c="orange")

sns.regplot(x='Climate: Average number of sunshine hours', y='Instagramability: Number of photos with #', data= dataframe,line_kws={"color":"blue"})

plt.title("Annual Sunshine vs Photos posted to Instagram")
```
